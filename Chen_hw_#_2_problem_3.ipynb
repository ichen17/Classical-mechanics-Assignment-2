```python
import numpy as np
import matplotlib.pyplot as plt
import brachis as br

def f(x):
    return -np.sin(np.pi*x/2.0)

def f_d(x):
    return -np.pi*np.cos(np.pi*x/2.0)/2.0

y=np.linspace(0,1,101)
plt.figure(figsize=(4,3),dpi=98)
plt.plot(y,f(y))
plt.annotate('Zero Second Drivative', xy=(0.0, 0.0), xytext=(0.2, -0.1),
             arrowprops=dict(facecolor='black', shrink=0.10),
             )

plt.grid(True)
plt.legend()
plt.show()

time=0.0
time_err=0.0
crvs = [f]
crvs_d = [f_d]
time, time_err=br.time_end(f, f_d,1)
print('curve sine takes a time = ', time, ' with error ', time_err)
    
```
